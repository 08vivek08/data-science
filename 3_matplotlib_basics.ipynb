```
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x=[1,5,4]
y=[4,5,1]
plt.plot([1,2,4],[2,5,7], label="first", linewidth=3)
plt.plot(x,y, label="second")
plt.legend() # to show label names wrt their color

plt.title("My data's graph")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.show()
```

```
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x=[1,5,4]
plt.scatter([1,2,4],[2,5,7], label="first", linewidth=3)
plt.hist(x, label="second")
plt.legend() # to show label names wrt their color

plt.title("My data's graph")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.show()
```
